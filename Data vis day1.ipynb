```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [0,2,4,6,8]

plt.figure(figsize=(5,4), dpi=100)

plt.plot(x,y, label='2x')
plt.title('Line graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.savefig('mygraph.png', dpi=300)

plt.legend()
plt.show()
```
