```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.countplot('species', data=df)
```

```python
sns.countplot(y='species', data=df)
```

```python
sns.barplot(x='species', y='petal_width', data=df)
```

```python
sns.barplot(x='species', y='sepal_length', data=df)
```

```python

sns.boxplot(y = 'petal_length', x= 'species', data = df)
```

```python
sns.boxplot(y = 'sepal_length', x= 'species', data = df, palette="rainbow")
```

```python
sns.boxplot(x='petal_length', y='species', hue = 'species', data = df, orient="h")
```

```python
sns.violinplot(x='sepal_width', y = 'species', data = df)
```
